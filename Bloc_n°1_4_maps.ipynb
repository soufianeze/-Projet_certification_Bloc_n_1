```python
df_meteo = pd.read_csv("df_prev_meteo.csv",index_col=[0] )
```

```python
# Déterminer la température minimale et la température maximale
tempmax = df_meteo['temperature'].max()
tempmin = df_meteo['temperature'].min()
```

```python
# Catographies des villes
fig = px.scatter_mapbox(df_meteo, lat = 'latitude', lon = 'longitude',animation_frame = 'jour', hover_name = 'villes',color='temperature', hover_data= {
                            "temperature": True,   
                            "météo": True,
                            "latitude": False,
                            "longitude": False,
                            'jour': False},
                            color_continuous_scale="thermal", range_color=[tempmin,tempmax],
                         title='Météo et prévision 7 jours',
                        mapbox_style="carto-positron", width = 800, height=800, zoom=4
                        )
fig.update_traces(marker={'size': 18})
fig.show()
```
